```python
import geopandas as gpd
import pandas as pd
```

```python
cities = gpd.read_file("../data/ne_10m_populated_places_simple.geojson")
```

```python
eclipse_2024 = gpd.read_file("../data/upath_lo.geojson")
```

```python
cities_clip = gpd.clip(cities, eclipse_2024)
```

```python
cities_clip = cities_clip.rename(columns={"name": "Name", "pop_max": "Population"})
```

```python
cities_clip.to_file("../data/cities_eclipse_2024.geojson")
```

```python
cities_clip[["Name", "Population"]].sort_values(by="Population", ascending=False).head(5)
```

```python
m = eclipse_2024.explore(tiles="CartoDB positron", tooltip=False, style_kwds={"stroke":True, "color": "yellow", "fillColor":"gray", "fillOpacity":0.6})
cities_clip.explore(m=m, column="Population", cmap="winter_r", tooltip_kwds={"sticky": True}, legend_kwds = {"colorbar": True, "max_labels":-2}, tooltip=["Name", "Population"], popup=["Name", "Population"], style_kwds={"style_function":lambda x: {"radius":(x["properties"]["Population"])/300000}}, legend=True)
```

```python
m.save("../eclipse_2024_cities.html")
```
